### Importaciones

```python
import pandas as pd
import matplotlib.pyplot
```

importamos el modelo de rgresion y los reguladores Lasso y Ridge

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
```

importamos la metrica Error Cuadratico Medio

```python
from sklearn.metrics import mean_squared_error
```

Importamos la funcion para dividir el datasets en 2 partes

```python
from sklearn.model_selection import train_test_split
```

### Creamos el Df

```python
df = pd.read_csv("whr2017.csv", index_col=False)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

### Dividimos en features y target

```python
X = df.drop(columns=["score", "country"])
Y = df[["score"]]
```

### Divimos en conjunto de train y test

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

### Crear el modelo de regresion lineal

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

Aplicacion la metrica del error cuadratico medio para mejor la distancia entre la prediccion del modelo y los datos reales

EL número que nos devuelve, nos dira que tan cerca o lejos esta nuestro modelo de las predicciones reales. CUAndo mas cerca a 0 este, mejor son las predicciones.

```python
Y_predic = model.predict(X_test)

mse = mean_squared_error(Y_test, Y_predic)

print("mse", mse)

if mse > 0 and mse < 1:
    print("El modelo es buenisimo")
else:
    print("Falta regularizar el modelo")
```

### Ahora, aplicamos el REGULADOR LASSO

Lasso (Least Absolute Shrinkage and Selection Operator) es una técnica de regresión lineal que utiliza regularización L1. Esto significa que, además de minimizar el error cuadrático entre las predicciones y los valores reales, Lasso añade una penalización proporcional al valor absoluto de los coeficientes del modelo. 

El efecto principal de esta penalización es que puede reducir algunos coeficientes exactamente a cero, lo que equivale a seleccionar automáticamente las variables más importantes y eliminar las menos relevantes. Por eso, Lasso es útil no solo para mejorar la generalización del modelo y evitar el sobreajuste, sino también para realizar selección de variables.

le pasamos como parametro alpha=0.02, es la cantidad de penalizaciones que recibiran nuestras features

```python
model_lasso = Lasso(alpha=0.02).fit(X_train, Y_train)
Y_predic_lasso = model_lasso.predict(X_test)
```

Probamos la metrica MSE con nuestro modelo Lasso 

```python
mse_lassso = mean_squared_error(Y_test, Y_predic_lasso)

print("mse", mse_lassso)

if mse_lassso > 0 and mse_lassso < 1:
    print("El modelo es buenisimo")
else:
    print("Falta regularizar el modelo")
```

Ahora probaremos el modelo con regularizador Ridge

Ridge es una técnica de regresión lineal que utiliza regularización L2. Esto significa que, además de minimizar el error cuadrático entre las predicciones y los valores reales, Ridge añade una penalización proporcional al cuadrado de los coeficientes del modelo.

El efecto principal de esta penalización es reducir el valor absoluto de los coeficientes, pero sin llevarlos exactamente a cero (a diferencia de Lasso). Ridge es útil para evitar el sobreajuste, especialmente cuando hay muchas variables correlacionadas, ya que distribuye el peso entre ellas en lugar de eliminar algunas por completo.

```python
model_ridge = Ridge(alpha=1).fit(X_train, Y_train)
y_predic_ridge = model_ridge.predict(X_test)
```

APlicamos MSE a nuestro modelo Ridge

```python
mse_redge = mean_squared_error(Y_test, y_predic_ridge)

print("mse", mse_redge)

if mse_redge > 0 and mse_redge < 1:
    print("El modelo es buenisimo")
else:
    print("Falta regularizar el modelo")
```

En este caso, cual es mejor? 

```python
print("mse sin reguladores", mse)
print("mse con lasso", mse_lassso)
print("mse con redge", mse_redge)
```

Los regularizadores L1 (Lasso) y L2 (Ridge) se utilizan en modelos de regresión para prevenir el sobreajuste.

    L1 (Lasso): Es conveniente cuando se espera que solo un pequeño número de características sea relevante. Este método puede eliminar completamente algunas variables al reducir sus coeficientes a cero, facilitando la selección de características.

    L2 (Ridge): Se recomienda cuando se tiene un gran número de variables que podrían ser relevantes, ya que tiende a distribuir el peso entre todas las características, evitando que algunos coeficientes se vuelvan cero.

La elección depende de la naturaleza de tus datos y el objetivo del modelo.
